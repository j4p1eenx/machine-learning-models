```python
import pandas as pd
df = pd.read_csv(r"C:\Users\kjapl\Downloads\titanic.csv")
df.head()
```

```python
df_n=df.drop(['Name','SibSp','Parch','Ticket','Cabin','Embarked','Survived','PassengerId'],axis='columns')
target=df['Survived']
df_n
```

```python
df_n.Sex = df_n.Sex.map({'male':1,'female':2})
df_n
```

```python
df_n.Age[:10]
```

```python
df_n.Age=df_n.Age.fillna(df_n.Age.mean())
df_n
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df_n,target,test_size=0.3)
```

```python
model.fit(X_train,y_train)
model.score(X_test,y_test)
```
